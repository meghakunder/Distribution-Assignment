**4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.**

**Ans:**

Standardization is the process of converting raw data values into Z-scores, which show how far each value is from the mean in terms of standard deviations. It allows different variables to be compared on the same scale, even if they originally had different units or ranges.

**Standardization Is Important in Data Analysis because**:

* Puts all variables on the same scale
 After standardization, the mean becomes 0 and the standard deviation becomes  This helps compare features that originally had different units (e.g., height in cm, weight in kg).

* Helps detect outliers
 Z-scores show how far a point is from the mean.
Z > 3 or Z < 3  possible outlier.

* Required for many statistical models and machine learning  algorithmsAlgorithms like regression, clustering, PCA, and distance-based  models work best when data is standardized.

* Improves model performance
 Variables with large scales do not dominate smaller variables.

 Z-Score Formula

	z=X/


Where:
X = original data value

 = mean of the dataset

 = standard deviation

This formula subtracts the mean from each value and then divides by the standard deviation.

**Standardization Transforms a Dataset**

The new mean becomes 0

The new standard deviation becomes 1

The overall shape of the distribution stays the same

Values become relative to the mean instead of absolute

Example:
A Z-score of +2 means the value is 2 standard deviations above the mean, and a Z-score of 1 means it is 1 standard deviation below the mean.

**5.What is Kurtosis and their type?**

**Ans:**

Kurtosis is a measure that helps us understand the shape of a distribution. It shows whether the data is sharply peaked or more flat, and whether it contains many extreme values (outliers). In simple terms, kurtosis tells us how concentrated the data is around the center and how heavy or light the tails of the distribution are.

**Types of Kurtosis**
1. Mesokurtic (Normal Kurtosis)

Represents the shape of the standard normal distribution.

Exhibits moderate peak and moderate tails.

Indicates a balanced distribution without extreme outliers.

Kurtosis value: approximately 3

2. Leptokurtic (High Kurtosis)

Has a sharper, higher peak than a normal distribution.

Shows heavier tails, meaning more extreme values or outliers.

Data is tightly clustered around the mean but also contains more extreme deviations.

Kurtosis value: greater than 3

3. Platykurtic (Low Kurtosis)

Features a flatter, wider peak compared to a normal distribution.

Displays lighter tails, meaning fewer extreme values.

Data is more spread out and less concentrated around the mean.

Kurtosis value: less than 3



**6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.**

**Ans:**

A fair die has six possible outcomes:
1, 2, 3, 4, 5, 6

Each of these outcomes has the same chance of occurring, which is:

P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=1/6

This means:

No number is more likely than another

All outcomes are equally likely

That is exactly the condition of a uniform distribution.

The uniform distribution is a good model for a fair die because all six outcomes have equal probability, so each result is equally likely.

**8.How does log transformation help in stabilizing variance and making data more normally distributed?**

**Ans:**

A log transformation helps when the data has very large values or is very uneven. It does two main things:

**Stabilizes Variance**

In many datasets, big numbers vary a lot more than small numbers.

When we take the log, large values become smaller, so the data becomes more balanced.

This makes the spread of the data more even.

**Makes Data More Normal**

Sometimes data is very skewed (long tail on the right), like income or sales.

Taking the log pulls in the long tail and makes the data more symmetric.

This makes the data look more like a normal bell-shaped curve.
